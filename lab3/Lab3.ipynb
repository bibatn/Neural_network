```python
from sklearn.datasets import make_blobs
from sklearn import metrics
import matplotlib.pyplot as plt
import numpy as np
num_classes=4
```

# Пример 1
линейно разделимые множества (с расстоянием между группами в 10^3 раз больше, чем диаметр группы)

```python
X1, Y1 = make_blobs(n_samples = 1000, n_features=2, centers = 4, cluster_std = 0.001, random_state = 10)
plt.scatter(X1[:,0],X1[:,1], c = Y1, s=50, cmap = 'magma')
print(X1.shape)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=num_classes, random_state=1).fit(X1)
labels = kmeans.labels_
```

```python
print("Fowlkes-Mallows scores:", metrics.fowlkes_mallows_score(Y1, labels))
print("Homogeneity_score:", metrics.homogeneity_score(Y1, labels))
print("Adjusted Rand index:", metrics.adjusted_rand_score(Y1, labels))
print("Adjusted Mutual Information:", metrics.adjusted_mutual_info_score(Y1, labels))
print("V-measure:", metrics.v_measure_score(Y1, labels))
plt.scatter(X1[:,0],X1[:,1], c=labels, s=50, cmap='magma')

plt.show()
```

```python
from sklearn.cluster import SpectralClustering
SC = SpectralClustering(n_clusters=num_classes, random_state=1).fit(X1)
labels = SC.labels_
```

```python
print("Fowlkes-Mallows scores:", metrics.fowlkes_mallows_score(Y1, labels))
print("Homogeneity_score:", metrics.homogeneity_score(Y1, labels))
print("Adjusted Rand index:", metrics.adjusted_rand_score(Y1, labels))
print("Adjusted Mutual Information:", metrics.adjusted_mutual_info_score(Y1, labels))
print("V-measure:", metrics.v_measure_score(Y1, labels))
plt.scatter(X1[:,0],X1[:,1], c=labels, s=50, cmap='magma')

plt.show()
```

```python
from sklearn_extra.cluster import KMedoids
KM = KMedoids(num_classes, metric = 'euclidean').fit(X1)
labels = KM.labels_
```

```python
print("Fowlkes-Mallows scores:", metrics.fowlkes_mallows_score(Y1, labels))
print("Homogeneity_score:", metrics.homogeneity_score(Y1, labels))
print("Adjusted Rand index:", metrics.adjusted_rand_score(Y1, labels))
print("Adjusted Mutual Information:", metrics.adjusted_mutual_info_score(Y1, labels))
print("V-measure:", metrics.v_measure_score(Y1, labels))
plt.scatter(X1[:,0],X1[:,1], c=labels, s=50, cmap='magma')

plt.show()
```

```python
from sklearn_extra.cluster import KMedoids
KM = KMedoids(num_classes, metric = 'manhattan').fit(X1)
labels = KM.labels_
```

```python
print("Fowlkes-Mallows scores:", metrics.fowlkes_mallows_score(Y1, labels))
print("Homogeneity_score:", metrics.homogeneity_score(Y1, labels))
print("Adjusted Rand index:", metrics.adjusted_rand_score(Y1, labels))
print("Adjusted Mutual Information:", metrics.adjusted_mutual_info_score(Y1, labels))
print("V-measure:", metrics.v_measure_score(Y1, labels))
plt.scatter(X1[:,0],X1[:,1], c=labels, s=50, cmap='magma')

plt.show()
```

# Пример 5


```python
import pandas as pd
import copy
```

```python
col_names = ['buying','maint','doors', 'persons', 'lug_boot', 'safety','class']
```

```python
df = pd.read_csv("/Users/M/Desktop/Нейронки/lab_1/car.data", names=col_names)
```

```python
buying_label = { ni: n for n, ni in enumerate(set(df['buying']))}
maint_label = { ni: n for n, ni in enumerate(set(df['maint']))}
doors_label = { ni: n for n, ni in enumerate(set(df['doors']))}
persons_label = { ni: n for n, ni in enumerate(set(df['persons']))}
lug_boot_label = { ni: n for n, ni in enumerate(set(df['lug_boot']))}
safety_label = { ni: n for n, ni in enumerate(set(df['safety']))}
class_label = { ni: n for n, ni in enumerate(set(df['class']))}
```

```python
df1 = copy.deepcopy(df)
```

```python
df1['buying'] = df1['buying'].map(buying_label)
df1['maint'] = df1['maint'].map(maint_label)
df1['doors'] = df1['doors'].map(doors_label)
df1['persons'] = df1['persons'].map(persons_label)
df1['lug_boot'] = df1['lug_boot'].map(lug_boot_label)
df1['safety'] = df1['safety'].map(safety_label)
df1['class'] = df1['class'].map(class_label)
```

```python
Xfeatures = df1[['buying','maint','doors', 'persons', 'lug_boot', 'safety']]
Ylabels = df1['class']
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=1).fit(Xfeatures)
labels = kmeans.labels_
```

```python
print("Fowlkes-Mallows scores:", metrics.fowlkes_mallows_score(Ylabels, labels))
print("Homogeneity_score:", metrics.homogeneity_score(Ylabels, labels))
print("Adjusted Rand index:", metrics.adjusted_rand_score(Ylabels, labels))
print("Adjusted Mutual Information:", metrics.adjusted_mutual_info_score(Ylabels, labels))
print("V-measure:", metrics.v_measure_score(Ylabels, labels))
```

```python
from sklearn.cluster import SpectralClustering
SC = SpectralClustering(n_clusters=num_classes, random_state=1).fit(Xfeatures)
labels = SC.labels_
```

```python
print("Fowlkes-Mallows scores:", metrics.fowlkes_mallows_score(Ylabels, labels))
print("Homogeneity_score:", metrics.homogeneity_score(Ylabels, labels))
print("Adjusted Rand index:", metrics.adjusted_rand_score(Ylabels, labels))
print("Adjusted Mutual Information:", metrics.adjusted_mutual_info_score(Ylabels, labels))
print("V-measure:", metrics.v_measure_score(Ylabels, labels))
```
